```python
import numpy as np
import pandas as pd

# freeze the random seed
np.random.seed(424)
```

```python
# generate the original label
label_df = pd.DataFrame()
label_df['Original_Label'] = pd.Series(np.random.randint(1, 101, size=10))
```

```python
# Log Label
label_df['Log10_Label'] = label_df['Original_Label'].apply(np.log10)

# two-class label
label_df['TwoCalss_label'] = label_df['Original_Label'].apply(lambda x: 1 if x > 50 else 0)

# multi-class label
condictions = [label_df['Original_Label']<=30,
               (label_df['Original_Label']>30)&(label_df['Original_Label']<80),
               label_df['Original_Label']>=80]
values = [0, 1, 2]
label_df['ThreeClass_label'] = np.select(condictions, values)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Import Data
===

```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
print('shape of the car crash: {}'.format(crash_df.shape))
crash_df.head()
```

Distribution Plots
===
Distribution Plot (1-variable)
---

```python
sns.displot(crash_df['not_distracted'], kde=True, bins=12)
```

Joint Plot (2-variables)
---

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

KDE Plot
---

```python
sns.kdeplot(crash_df['alcohol'])
```

Pair Plots
---

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')

print('shape of the tips: {}'.format(tips_df.shape))
tips_df.head()
```

```python
sns.pairplot(data=tips_df, hue='sex', palette='Blues')
```

Rug Plot
---

```python
sns.rugplot(tips_df['tip'])
```

Styling
---

```python
sns.set_style('ticks')
plt.figure(figsize=(8, 4))
sns.set_context('paper', font_scale=1.5)
# sns.despine(left=True) 
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
sns.despine(left=True, bottom=True, right=True, top=True)
```

Categorical Plots
===
Bar Plot
---

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.var)
```

Count Plot
---

```python
sns.countplot(x='sex', data=tips_df)
```

Box Plot
---

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')

plt.legend(loc=(1.01,0.4))
```

Violin Plot
---

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)

plt.legend(loc=(1.01,0.4))
```

Strip Plot
---

```python
plt.figure(figsize=(8,5))
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)
plt.legend(loc=(1.01,0.4))
```

Swarm Plot
---

```python
sns.swarmplot(x='day', y='total_bill', data=tips_df)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='white')
```

Palettes
---

```python
plt.figure(figsize=(8,6))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='afmhot')
plt.legend(loc=(1.01,0.4))
```

Matrix Plots

Heatmaps
---

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)

crash_mx = crash_df.drop(columns='abbrev').corr()
sns.heatmap(crash_mx, annot=True, cmap='Blues')
```

```python
flights_df = sns.load_dataset('flights')
print('shape of flights: {}'.format(flights_df.shape))
flights_df.head()
```

```python
flights = flights_df.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidths=1)
```

PairGrid
---

```python
iris_df = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris_df, hue='species')
# iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
# iris_g.map_offdiag(plt.scatter)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

```python
iris_g = sns.PairGrid(iris_df, hue='species', x_vars=['sepal_length', 'sepal_width'], y_vars=['petal_length', 'petal_width'])
iris_g.map(sns.scatterplot)
# iris_g.map(plt.scatter)
iris_g.add_legend()
```

Facet Grid
---

```python
tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
tips_fg.map(plt.hist, 'total_bill', bins=8)
# tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.3, col_order=['Dinner', 'Lunch'], palette='Set1')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
```

```python
kws = dict(s=50, linewidth=.5, edgecolor='w')
tips_fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', height=4, aspect=1.3, hue_order=['Yes', 'No'], hue_kws=dict(marker=['^', 'v']))
tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
```

```python
att_df = sns.load_dataset('attention')

print('shape of the attention: {}'.format(att_df.shape))
att_df.head()
```

```python
att_fg = sns.FacetGrid(att_df, col='subject', col_wrap=5, height=1.5)
att_fg.map(plt.plot, 'solutions', 'score', marker='.')
```
